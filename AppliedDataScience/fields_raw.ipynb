```python
from pathlib import Path
import pandas as pd
for col in df2.columns:
    info = core_info.copy()
    info.append(col)
    col_df = df.loc[:, info]
    Path('fields_csv').mkdir(exist_ok=True)
    col_df.to_csv(f'fields_csv/{col}.csv', index=False)
```

```python
import os

def create_full_directory_structure(base_path):
    # List of subdirectories to create
    subdirectories = [
        "Capstone或Thesis",
        "面试",
        "其他字段",
        "课程网址",
        "职业项目",
        "项目标签",
        "所属院系（英文）",
        "托福送分ETS code",
        "推荐信",
        "更多项目信息",
        "申请费减免"
    ]
    
    # Create the base directory
    os.makedirs(base_path, exist_ok=True)
    
    # Create each subdirectory
    for subdir in subdirectories:
        os.makedirs(os.path.join(base_path, subdir), exist_ok=True)
    
    print(f"Directories created under: {base_path}")

# Example usage
create_full_directory_structure('/Users/yijingyang/Library/CloudStorage/OneDrive-个人/GradPilot/ProgramDB/AppliedDataScience/fields_records')
```
